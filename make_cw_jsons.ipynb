```python
genre_counts = all_albums['genre'].value_counts()
```

```python
cw_dict = {}
cw_year_dict = {}
num_cw = 11

for key,row in all_albums.iterrows():
    for i in range(1,num_cw):
        genre_cw = row['genre'] + '_' + str(row['cw' + str(i)])
        if genre_cw not in cw_dict.keys():
            cw_dict[genre_cw] = row['cwc' + str(i)]
        else:
            cw_dict[genre_cw] += row['cwc' + str(i)]
            
        year_genre_cw = str(row['year']) + '_' + row['genre'] + '_' + str(row['cw' + str(i)])
        if year_genre_cw not in cw_year_dict.keys():
            cw_year_dict[year_genre_cw] = row['cwc' + str(i)]
        else:
            cw_year_dict[year_genre_cw] += row['cwc' + str(i)]
```

```python
cw_dict
```

```python
cw_dict_list = []
for key,value in cw_dict.items():
    curr_dict = {}
    curr_dict['genre'] = key.split('_')[0]
    curr_dict['word'] = key.split('_')[1]
    curr_dict['freq'] = value
    curr_dict['freq_norm'] = value/genre_counts[key.split('_')[0]]
    cw_dict_list.append(curr_dict)
```

```python
cw_year_dict
```

```python
cw_year_dict_list = []
for key,value in cw_year_dict.items():
    curr_dict = {}
    curr_dict['year'] = int(key.split('_')[0])
    curr_dict['genre'] = key.split('_')[1]
    curr_dict['word'] = key.split('_')[2]
    curr_dict['freq'] = value
    cw_year_dict_list.append(curr_dict)
```

```python
import json

with open('overall_common_word.json', 'w') as f:
    json.dump(cw_dict_list, f, indent=2)
    print("New json file is created from data.json file")
    
with open('yearly_common_word.json', 'w') as f:
    json.dump(cw_year_dict_list, f, indent=2)
    print("New json file is created from data.json file")
```
